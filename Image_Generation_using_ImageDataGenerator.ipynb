```python

from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rotation_range=40,horizontal_flip=0.2,height_shift_range=0.2,
                                   brightness_range=[0.7,0.99],shear_range=0.2,channel_shift_range=100,
                                  width_shift_range=0.2,zoom_range=0.2,vertical_flip=0.2)

data_train = 'C:/Users/Administrator/Desktop/nn_scratch/images'

aug_dir = 'C:/Users/Administrator/Desktop/nn_scratch/gen_imgs/females'
```

```python
train_generator = train_datagen.flow_from_directory(
data_train, 
target_size=(256,256), 
batch_size=100,
save_to_dir=aug_dir,
save_format='jpg'
)
```

```python
def plot_images(train_generator):
    plt.figure(figsize=(18,6))
    for i in range(10):
        plt.subplot(2,5, i+1)
        batch =train_generator.next()
        image = batch[0].astype('uint8')
        plt.imshow(image[0])
        plt.axis('off')
    plt.show()
plot_images(train_generator)
```
